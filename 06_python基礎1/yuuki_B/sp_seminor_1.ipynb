```python
import matplotlib.pyplot as plt
import numpy as np
import math
```

```python
def sigmoid(x):
  return 1/(1+np.exp(-x))

def softmax(x):
  exp_x = np.exp(x)
  sum_exp_x = np.sum(exp_x)
  return exp_x / sum_exp_x

def tanh(x):
 return (np.exp(x)-np.exp(-x) ) / (np.exp(x) + np.exp(-x))

x = np.arange(-5.0,5.0,0.1)

tan = tanh(x)
plt.ylim(-1.2,1.2)
plt.plot(x,tan,label ='tanh')
plt.title('work1')
plt.xlabel('input')
plt.ylabel('output')

plt.grid()
plt.legend()
plt.show()
```

```python

x = np.arange(-5.0,5.0,0.1)

sig = sigmoid(x)
soft = softmax(x)
tan = tanh(x)




plt.plot(x,sig,label ='sigmoid')
plt.plot(x,soft,label ='softmax')
plt.plot(x,tan,label ='tanh')

plt.ylim(-1.2,1.2)
plt.title('work2')
plt.xlabel('input')
plt.ylabel('output')

plt.grid()
plt.legend()
plt.show()
```

```python

from sympy.geometry import *
import matplotlib.pyplot as plt
%matplotlib inline

tmp = ((1,1),(2,3),(3,1))
p0 = Polygon(*tmp)

tmp = ((2,3),(3,5),(4,3))
p1 = Polygon(*tmp)

tmp= ((3,1),(4,3),(5,1))
p2 = Polygon(*tmp)

polygons = [p0,p1,p2]

plt.figure(figsize=(5,5),dpi=96)

for p in polygons :
    tmp = list(map(lambda p:(float(p.x),float(p.y)), p.vertices))
    tmp = plt.Polygon(tmp,fc='yellow',alpha=1.0)
    ax = plt.gca()
    ax.add_patch(tmp)
    ax.set_facecolor('black')


xRange=yRange=(0,6) 
plt.xlim(xRange)
plt.ylim(yRange)
plt.xticks(range(xRange[0],xRange[1]+1))
plt.yticks(range(yRange[0],yRange[1]+1))
#plt.grid(color='blue', alpha=0.3)
plt.show()
```

課題2　3つの活性化関数をまとめて表示してみよう．
> 条件：x = np.arange(-5.0, 5.0, 0.1)， plt.ylim(-1.2,1.2)，
凡例，軸ラベルの表示



```python
import matplotlib.pyplot as plt
import numpy as np

def sigmoid(x):
  return 1/(1+np.exp(-x))

def softmax(x):
  exp_x = np.exp(x)
  sum_exp_x = np.sum(exp_x)
  return exp_x / sum_exp_x

def tanh(x):
  molecular = np.exp(x) - np.exp(-x)
  denominator = np.exp(x) + np.exp(-x)
  return molecular / denominator

x = np.arange(-5.0,5.0,0.1)
y_sig = sigmoid(x)
y_soft = softmax(x)
y_tanh = tanh(x)

plt.plot(x,y_sig,label='sigmoid')
plt.plot(x,y_soft,label='softmax')
plt.plot(x,y_tanh,label='tanh')
plt.ylim(-1.2,1.2)

plt.title('activation function : sigmoid, softmax, tanh') #タイトル
plt.grid() #グリッドの表示
plt.legend() #凡例の表示
plt.xlabel('input') #x軸ラベル
plt.ylabel('output') #y軸ラベル
plt.show()
```

**課題4**　ゼルダのトライフォースを作ってみよう．

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

image = np.zeros((300,300,3),dtype = np.uint8)  #背景(黒)

#left
pt1 = (100, 150)
pt2 = (50, 250)
pt3 = (150, 250)

#right
pt4 = (200, 150)
pt5 = (150, 250)
pt6 = (250, 250)

#top
pt7 = (150, 50)
pt8 = (100, 150)
pt9 = (200, 150)

triangle_cnt = np.array( [pt1, pt2, pt3] )
cv2.drawContours(image, [triangle_cnt], 0, (250,240,10), -1)
triangle_cnt = np.array( [pt4, pt5, pt6] )
cv2.drawContours(image, [triangle_cnt], 0, (250,240,10), -1)
triangle_cnt = np.array( [pt7, pt8, pt9] )
cv2.drawContours(image, [triangle_cnt], 0, (250,240,10), -1)
plt.imshow(image)
plt.axis('off')
```

```python
import matplotlib.pyplot as plt
import numpy as np
import cv2

img = cv2.imread('blak.jpg')
#plt.axis('off')

pts1 = np.array([(80, 125), (125, 50), (170, 125)])
pts2 = np.array([(80, 125), (125, 200), (35, 200)])
pts3 = np.array([(215, 200), (125, 200), (170, 125)])

cv2.fillPoly(img, [pts1, pts2, pts3], (255, 255, 0))

plt.imshow(img)
cv2.destroyAllWindows()
```
